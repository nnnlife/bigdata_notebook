```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
from scipy.stats import chi2

# 1. Create a sample dataset for binary classification
X, y = make_classification(n_samples=100, n_features=5, random_state=42, 
                           n_informative=3, n_redundant=0)
df = pd.DataFrame(X, columns=['X1', 'X2', 'X3', 'X4', 'X5'])
df['y'] = y

# 2. Fit a Logistic Regression Model using statsmodels for p-values and coefficients
X = sm.add_constant(df[['X1', 'X2', 'X3', 'X4', 'X5']])  # Add intercept
model = sm.Logit(df['y'], X)
result = model.fit()

# Display summary to understand the p-values and coefficients
print(result.summary())

# 3. Identify the number of insignificant variables
# Assuming alpha = 0.05 for significance level
p_values = result.pvalues[1:]  # Skip the intercept
insignificant_vars = p_values[p_values > 0.05].index
num_insignificant_vars = len(insignificant_vars)
print(f"Number of insignificant variables: {num_insignificant_vars}")

# 4. Calculate sum and mean of coefficients for only significant variables
significant_vars = p_values[p_values <= 0.05].index
significant_coefficients = result.params[significant_vars]
coeff_sum = significant_coefficients.sum()
coeff_mean = significant_coefficients.mean()
print(f"Sum of significant coefficients: {coeff_sum}")
print(f"Mean of significant coefficients: {coeff_mean}")

# 5. Calculate the Odds Ratio for a variable increasing by 5 units
# Let's assume we want to calculate this for 'X1'
odds_ratio = np.exp(result.params['X1'] * 5)
print(f"Odds Ratio for X1 increasing by 5 units: {odds_ratio}")
```
